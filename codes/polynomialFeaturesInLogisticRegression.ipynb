```python
from sklearn.preprocessing import PolynomialFeatures
pol = PolynomialFeatures(degree=3, include_bias=False)
X_trf = pol.fit_transform(X_train)
```

```python
clf1 = LogisticRegression(multi_class='multinomial')
clf1.fit(X_trf, y_train)
np.mean(cross_val_score(clf1, X_trf, y_train, scoring='accuracy', cv=10))
```

```python
def plot_decision_boundary(X_train, y_train, degree=1): 

    pol = PolynomialFeatures(degree=degree)
    X_trf = pol.fit_transform(X_train)

    clf = LogisticRegression()
    clf.fit(X_trf, y_train)

    accuracy = np.mean(cross_val_score(clf, X_trf, y_train, scoring='accuracy', cv=10))

    a = np.arange(start=X_train[:,0].min()-1, stop=X_train[:, 0].max()+1, step=0.01)
    b = np.arange(start=X_train[:,1].min()-1, stop=X_train[:, 1].max()+1, step=0.01)

    XX, YY = np.meshgrid(a, b)

    input_array = np.array([XX.ravel(), YY.ravel()]).T

    labels = clf.predict(pol.transform(input_array))

    plt.contour(XX, YY, labels.reshape(XX.shape), alpha=0.5)
    plt.scatter(X_train[:,0], X_train[:,1], c=y_train)
    plt.title('Degree = {}, accuracy is {}'.format(degree, np.round(accuracy,4)))
```

```python
plot_decision_boundary(X_train, y_train)
```

```python
plot_decision_boundary(X_train, y_train, degree=2)
```

```python
plot_decision_boundary(X_train, y_train, degree=3)
```

```python
plot_decision_boundary(X_train, y_train, degree=4)
```
