```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tested.csv')
```

```python
df.head()
```

# 1 Categorical Data

## a. Countplot

```python
# sns.countplot(df['Pclass'])
df['Survived'].value_counts().plot(kind='bar')
```

## b. PieChart

```python
df['Pclass'].value_counts().plot(kind='pie', autopct = '%.2f')
```

```python
sns.distplot(df['Age'])
```

```python
sns.boxplot(df['Fare'])
```
